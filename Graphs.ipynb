```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
#output the graph in same shell
```

```python
iris=sns.load_dataset('iris')
sns.jointplot(x='petal_length',y='petal_width',data=iris)
```

```python
#Facet grid plot - variable in data to map plot-aspects in different colors
sns.FacetGrid(iris,hue='species',size=5)\
.map(plt.scatter,'sepal_length','sepal_width')\
.add_legend()
```

```python
#bar graph

sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
#violin plot

sns.violinplot(x='species',y='petal_length',data=iris)
```

```python
#pair plot

sns.pairplot(data=iris,kind='scatter')
```

```python
#heat map

df=pd.DataFrame(np.random.rand(7,5))
#print(df)
corr=df.corr()
print(corr)
sns.heatmap(corr)
```

```python
flights=sns.load_dataset('flights')
#print(flights)
#flights.head()

flights=flights.pivot('month','year','passengers')
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True,fmt='d') #fmt gives values in described datatype such as d,f
#annot gives numerical value to each correlation
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidth=5,cmap='YlGnBu')
```

```python
#Dist plot

x=np.random.normal(size=100)
#print(x)
sns.distplot(x)
```

```python
sns.distplot(x,kde=False,rug=True)
#kde removes the density curve and rug gives the tweaks at the bottom
```

```python
sns.distplot(x,bins=5,kde=False,rug=True) #bins generalises the graph
```

```python
#kernel density estimation - useful for plotting the shape of a distribution

sns.distplot(x,hist=False,rug=True) #hist removes the histogram
```
